```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
from astropy.visualization import simple_norm
import matplotlib.colors as colors

filename = "../fit/agn.fits"
```

```python
sourceFile = "../psfConstruction/2020-02-21_J_J1215+1344_c1-4_58900_32019.mos.fits"
d = fits.getdata(sourceFile)
```

```python
with fits.open(filename) as hdul:
    f = hdul[0]
    data = f.data
    
plt.imshow(data)
plt.colorbar();
```

```python
norm = simple_norm(data, 'log')
mymap = mpl.colors.LinearSegmentedColormap.from_list('',['darkgrey','black','linen'])
mymap='magma_r'

fig, ax = plt.subplots(1,3, figsize=(12,4))
im0 = ax[0].imshow(np.nan_to_num(np.log(data*1000+1)/np.log(1001), nan=0),cmap=mymap)
im1 = ax[1].imshow(data, norm=norm,cmap=mymap)
im2 = ax[2].imshow(np.nan_to_num(np.log(data),nan=-1),cmap=mymap)

[fig.colorbar([im0,im1,im2][i], ax = ax[i], shrink=0.5) for i in range(3)];
fig.tight_layout();
```

```python
image = np.arange(65536).reshape((256, 256))

# Create an ImageNormalize object
norm = simple_norm(image, 'sqrt')

# Display the image
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
im = ax.imshow(image, origin='lower', norm=norm)
fig.colorbar(im)
```
