```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import emcee 
import pyimfit 
from astropy.io import fits
import corner
from IPython.display import Latex
import sys

plt.rcParams['image.cmap'] = 'Blues'
imageFile = "test.fits"
imageAGN = fits.getdata(imageFile)
```

```python
sersic = {'PA': [95, 0, 360], 'ell_bulge': [0.05,0,1], 'n': [2.5,0,4],
'I_e': [20,0,100], 'r_e': [5,0,50]}
sersic_dict = {'name': "Sersic", 'label': "galaxy 1 bulge", 'parameters': sersic}

expo = {'PA': [95,0,360], "ell": [0.45,0,1], "I_0": [90, "fixed"], "h" : [15,0,50]}
expo_dict = {"name":"Exponential", 'label': 'galaxy 1 disk', 'parameters':expo}

funcset_dict_1 = {'X0': [150.1,148,152], 'Y0': [149.5,148,152], 
                    'function_list': [sersic_dict,expo_dict]}

expo1 = {'PA': [22,0,360], "ell": [0.25,0,1], "I_0": [10,0,100], "h" : [20,0,50]}
expo1_dict = {"name":"Exponential", 'label': 'galaxy 2 ', 'parameters':expo1}

funcset_dict_2 = {'X0': [225,224,226], 'Y0': [181.7,180,183], 
                    'function_list': [expo1_dict]}

model0 = {'function_sets': [funcset_dict_1]}
model1 = {'function_sets': [funcset_dict_2]}
model2 = {'function_sets': [funcset_dict_1, funcset_dict_2]}

m0 = pyimfit.ModelDescription.dict_to_ModelDescription(model0)
m1 = pyimfit.ModelDescription.dict_to_ModelDescription(model1)
m2 = pyimfit.ModelDescription.dict_to_ModelDescription(model2)

f0 = pyimfit.Imfit(m0)
f1 = pyimfit.Imfit(m1)
f2 = pyimfit.Imfit(m2)
f = [f0,f1,f2];
```

```python
fig,ax = plt.subplots(1,3)
for i in range(3):
    im = ax[i].imshow(f[i].getModelImage((300,300)))
    cbar0 = fig.colorbar(im, ax=ax[i], shrink=0.2)
print("Simulation of galaxy 1(sersic+expo), galaxy 2 (expo), and galaxy 1+2")
fig.tight_layout();
```

# fitting with exact same model

```python
imagefit = f2.getModelImage()

f2.loadData(imagefit)
p_bestfit = f2.doFit().params
plt.imshow(f2.getModelImage() -imagefit)
plt.colorbar()
for i in range(len(f2.numberedParameterNames)):
    print(f"{f2.numberedParameterNames[i]}: {p_bestfit[i]:.3f}")
```

# fitting with slightly different guess

```python
sersic = {'PA': [105, 0, 180], 'ell_bulge': [0.05,0,1], 'n': [5,0.5,10],
'I_e': [20,0,100], 'r_e': [5,1,50]}
sersic_dict = {'name': "Sersic", 'label': "galaxy 1 bulge", 'parameters': sersic}

expo = {'PA': [85,0,180], "ell": [0.45,0,1], "I_0": [100, "fixed"], "h" : [15,0,50]}
expo_dict = {"name":"Exponential", 'label': 'galaxy 1 disk', 'parameters':expo}

funcset_dict_1 = {'X0': [140.1,0,300], 'Y0': [150,0,300], 
                    'function_list': [sersic_dict,expo_dict]}

expo1 = {'PA': [22,0,180], "ell": [0.25,0,1], "I_0": [10,0,100], "h" : [20,0,50]}
expo1_dict = {"name":"Exponential", 'label': 'galaxy 2 ', 'parameters':expo1}

funcset_dict_2 = {'X0': [180,0,300], 'Y0': [170,0,300], 
                    'function_list': [expo1_dict]}

model2 = {'function_sets': [funcset_dict_1, funcset_dict_2]}

m2 = pyimfit.ModelDescription.dict_to_ModelDescription(model2)

f2 = pyimfit.Imfit(m2)

f2.loadData(imagefit)
p_bestfit = f2.doFit().params
plt.imshow(f2.getModelImage() -imagefit)
plt.colorbar()
for i in range(len(f2.numberedParameterNames)):
    print(f"{f2.numberedParameterNames[i]}: {p_bestfit[i]:.3f}")
```
